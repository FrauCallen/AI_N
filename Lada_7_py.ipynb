```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

Побудувати графік функції y = 5x^3 − 4
тип лінії – “.”
розташування легенди (ліво верх) 
діапазон значень – у=(5:80),х=(-40:60)
колір лінії - #225533 
прозорість лінії – 76% 

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
x = np.linspace(-5, 5,80)

fig, ax = plt.subplots()

y=5*(x**3) - 4
ax.scatter(x, y, color='#225533',marker='.',alpha=0.76, label='y = 5(x^3) − 4') 

ax.set_xlim(-40, 60)
ax.set_ylim(5, 80)

ax.legend(loc=2)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Графік функції y = 5x^3 − 4')
ax.grid()
plt.show()
```

Функція Еклі 

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(X, Y)
Z = -20 * np.exp(-0.2 * np.sqrt(0.5 * (X**2 + Y**2)))- np.exp(0.5 * (np.cos(2 * np.pi * X) + np.cos(2 * np.pi * Y))) + np.e + 20
surf = ax.plot_surface(X, Y, Z, rstride=5, cstride=5, cmap=cm.jet)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Функція Еклі')

plt.show()
```

Побудувати стовпчикову діаграму (не менше 10 стовпчиків) "Рівень продажу гречки за 
останні 10 місяців", дані можна взяти довільні або знайти в мережі Інтернет. У діаграмі 
передбачити назви стовпчиків. Колір використовувати як для попереднього графіку. 

```python
months = ['Січ', 'Лют', 'Бере', 'Квіт', 'Трав', 'Черв', 'Лип', 'Серп', 'Вере', 'Жов']
sales = [150, 200, 180, 220, 50, 270, 300, 280, 260, 230]


plt.bar(months, sales, color='#225533')


plt.xlabel('Місяці')
plt.ylabel('Рівень продажу')

plt.title('Рівень продажу гречки за останні 10 місяців')

plt.show()


```
